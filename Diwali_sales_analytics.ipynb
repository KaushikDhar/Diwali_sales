```python
#sales in per age group 
ax=sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#to find which state has highest sales in over all
```

```python
Sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(28)
sns.set(rc={'figure.figsize':(25,5)}) #25 is the width and 5 is the height
sns.barplot(data=Sales_state,x='State',y='Orders')
```

```python
a=sns.countplot(data=df,x='Marital_Status',hue='Gender')
sns.set(rc={'figure.figsize':(3,3)})
for bars in a.containers:
    a.bar_label(bars)
```
